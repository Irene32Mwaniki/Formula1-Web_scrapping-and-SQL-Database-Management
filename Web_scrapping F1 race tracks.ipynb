```python
import requests
import pandas as pd
from io import StringIO

def get_specific_f1_circuit_table():
    try:
        url = "https://en.wikipedia.org/wiki/List_of_Formula_One_circuits"
        response = requests.get(url)
        response.raise_for_status()  # Check for request errors

        print("Response Status Code:", response.status_code)  
        
        # Use StringIO to wrap the response text
        html_content = StringIO(response.text)
        
        # Use pandas to read all HTML tables directly from the StringIO object
        tables = pd.read_html(html_content)

        # Check if there are enough tables and retrieve the specific table
        if len(tables) > 2:
            specific_table = tables[2]  # Get the third table (index 2)
            print("\nSuccessfully extracted Table 2!")
            return specific_table
        else:
            print("Less than 3 tables found on the webpage.")
            return None  # Return None if there are not enough tables

    except requests.exceptions.RequestException as e:
        print(f"Error fetching the webpage: {e}")
    except Exception as e:
        print(f"Error processing the DataFrame: {e}") 
    return None  # Return None if there was an error

# Run the function to get the specific table
f1_circuit_table = get_specific_f1_circuit_table()

if f1_circuit_table is not None:
    print(f1_circuit_table.head())  # Display the first few rows of the retrieved table
else:
    print("Failed to extract the table.")
```

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO


def scrape_tables(url): 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser') 
    tables = pd.read_html(StringIO(response.text))
    dataframes = [] 
    for i, table in enumerate(tables): 
        year = url.split('/')[-1].split('.')[0]
        table['Year'] = year 
        dataframes.append(table)
    return dataframes 


urls =['https://www.formula1.com/en/results/2010/races', 'https://www.formula1.com/en/results/2011/races', 'https://www.formula1.com/en/results/2012/races', 
'https://www.formula1.com/en/results/2013/races', 
'https://www.formula1.com/en/results/2014/races', 
'https://www.formula1.com/en/results/2015/races', 
'https://www.formula1.com/en/results/2016/races', 
'https://www.formula1.com/en/results/2017/races', 
'https://www.formula1.com/en/results/2018/races', 
'https://www.formula1.com/en/results/2019/races', 
'https://www.formula1.com/en/results/2020/races', 
'https://www.formula1.com/en/results/2021/races', 
'https://www.formula1.com/en/results/2022/races', 
'https://www.formula1.com/en/results/2023/races', 
'https://www.formula1.com/en/results/2024/races' 
]
all_data = []

for url in urls:
    year_data = scrape_tables(url)
    all_data.extend(year_data)
final_df = pd.concat(all_data, ignore_index=True)


print(final_df)


final_df.to_excel('f1_data_by_years.xlsx', index=False)



```
